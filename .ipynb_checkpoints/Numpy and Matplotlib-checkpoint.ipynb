```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

xx, yy = np.meshgrid(x, y)
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

xx, yy = np.meshgrid(x, y)
```

```python
print(xx.shape)
print(yy.shape)
```

```python
plt.pcolormesh(xx)
plt.colorbar()
plt.title("xx")
plt.xlabel("x index")
plt.ylabel("y index")
plt.show()
```

```python
r = np.sqrt(xx**2 + yy**2)
theta = np.arctan2(yy, xx)
```

```python
plt.pcolormesh(xx, yy, r, shading='auto')
plt.colorbar()
plt.title("r")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.pcolormesh(xx, yy, theta, shading='auto')
plt.colorbar()
plt.title("theta")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
q =  np.cos(4 * r)**2 + np.sin(4 * theta)**2

plt.pcolormesh(xx, yy, q, shading='auto')
plt.colorbar()
plt.title("quantity")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
T = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/T.npy')

S = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/S.npy')

P = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/P.npy')

date = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/date.npy')

lat = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/lat.npy')

lon = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/lon.npy')

level = np.load('/home/emf2229/.cache/pooch/182581396e4937a1fb09550d60cbc382-float_data.zip.unzip/float_data/levels.npy')
```

```python
print(T.shape)
print(S.shape)
print(P.shape)
print(date.shape)
print(lat.shape)
print(lon.shape)
print(level.shape)
```

```python
plt.plot(T, level)
plt.gca().invert_yaxis()
plt.title("Temperature Profiles")
plt.xlabel("Temperature")
plt.ylabel("Level")
plt.show()
```

```python
plt.plot(S, level)
plt.gca().invert_yaxis()
plt.title("Salinity Profiles")
plt.xlabel("Salinity")
plt.ylabel("Level")
plt.show()
```

```python
plt.plot(P, level)
plt.gca().invert_yaxis()
plt.title("Pressure Profiles")
plt.xlabel("Pressure")
plt.ylabel("Level")
plt.show()
```

```python
T_mean = np.mean(T, axis=1)
T_std = np.std(T, axis=1)

S_mean = np.mean(S, axis=1)
S_std = np.std(S, axis=1)

P_mean = np.mean(P, axis=1)
P_std = np.std(P, axis=1)
```

```python
plt.errorbar(T_mean, level, xerr=T_std)
plt.gca().invert_yaxis()
plt.title("Mean Temperature Profile")
plt.xlabel("Temperature")
plt.ylabel("Level")
plt.show()
```

```python
plt.errorbar(S_mean, level, xerr=S_std)
plt.gca().invert_yaxis()
plt.title("Mean Salinity Profile")
plt.xlabel("Salinity")
plt.ylabel("Level")
plt.show()
```

```python
plt.errorbar(P_mean, level, xerr=P_std)
plt.gca().invert_yaxis()
plt.title("Mean Pressure Profile")
plt.xlabel("Pressure")
plt.ylabel("Level")
plt.show()
```
