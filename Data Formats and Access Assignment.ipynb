```python
import json
import requests
# Here's a sample JSON structure similar to what APIs return
sample_json = '''
{
  "station": "USC00305800",
  "name": "New York Central Park",
  "location": {
    "latitude": 40.7789,
    "longitude": -73.9692
  },
  "observations": [
    {"date": "2023-01-01", "temperature": 32, "precipitation": 0.0},
    {"date": "2023-01-02", "temperature": 28, "precipitation": 0.5},
    {"date": "2023-01-03", "temperature": 35, "precipitation": 0.0},
    {"date": "2023-01-04", "temperature": 38, "precipitation": 0.2},
    {"date": "2023-01-05", "temperature": 41, "precipitation": 0.0}
  ]
}
'''

# Parse the JSON
data = json.loads(sample_json)

# Access nested data
print("Station:", data['station'])
print("Location:", data['location'])
print("First observation:", data['observations'][0])
```

```python
print("Date, Temperature")
for obs in data['observations']:
    date = obs["date"]
    temp = obs["temperature"]
    print(date, temp)
```

```python
total_temp = 0
count = 0

for obs in data["observations"]:
    total_temp += obs["temperature"]
    count += 1

avg_temp = total_temp / count
print(f"Average temperature: {avg_temp}°F")
```

```python
print("\nDays with precipitation:")
for obs in data["observations"]:
    if obs["precipitation"] > 0:
        print(obs["date"], "precip:", obs["precipitation"])
```

```python
lat = 40.7789
lon = -73.9692

points_url = f"https://api.weather.gov/points/{lat},{lon}"
resp = requests.get(points_url)

print("Status code:", resp.status_code)
points_data = resp.json()

forecast_url = points_data["properties"]["forecast"]
print("Forecast URL:", forecast_url)
```

```python
forecast_resp = requests.get(forecast_url)
forecast_data = forecast_resp.json()

periods = forecast_data["properties"]["periods"]
```

```python
print("Period, Temperature")

for p in periods:
    print(p["name"], p["temperature"])
```

```python
total_temp = 0
count = 0

for p in periods:
    total_temp += p["temperature"]
    count += 1

avg_temp = total_temp / count
print("Average temperature:", avg_temp, "F")
```

```python
print("\nPeriods with Precipitation:")

for p in periods:
    precip = p["probabilityOfPrecipitation"]["value"]

    if precip is not None and precip > 0:
        print(p["name"], "=", precip, "% chance")
```

```python
import requests

# OPeNDAP provides metadata in different formats
# We'll get basic info about a climate dataset

base_url = "http://iridl.ldeo.columbia.edu/expert/SOURCES/.NOAA/.NCEP/.CPC/.UNIFIED_PRCP/.GAUGE_BASED/.GLOBAL/.v1p0/.Monthly/.RETRO/.rain/dods"

# Get DDS (Dataset Descriptor Structure) - describes the structure
dds_url = base_url + ".dds"
response = requests.get(dds_url)

print("Dataset Structure:")
print(response.text[:500])  # Print first 500 characters
```

The dimensions are time (T), latitude (Y), and longitude (X). The main variable is rain.

```python
das_url = base_url + ".das"
das_response = requests.get(das_url)

print(das_response.text[:1000])
```

The dataset contains monthly precipitation data. It has monthly data from January 1st, 1960. It has full latitude and longitude data, so it is global coverage. The units of rainfall is millimeters per day.
