```python
def kelvin_to_celsius(K):
    return k - 273.15

def celsius_to_kelvin(C):
    return c + 273.15
```

```python
def celsius_to_fahrenheit(temp, input_unit="celsius"):
    if input_unit == "kelvin":
        temp = temp - 273.15
    return temp * 9/5 + 32
```

```python
print(celsius_to_fahrenheit(0, input_unit="celsius"))   
print(celsius_to_fahrenheit(50, input_unit="celsius")) 
print(celsius_to_fahrenheit(100, input_unit="celsius")) 
```

```python
def farenheit_to_celsius(temp, output_unit="celsius"):
    temp = (temp - 32) * 5/9 
    
    if output_unit == "kelvin":
        temp = temp + 273.15
        
    return temp
```

```python
print(farenheit_to_celsius(0, output_unit="celsius"))   
print(farenheit_to_celsius(50, output_unit="celsius")) 
print(farenheit_to_celsius(100, output_unit="celsius")) 
```

```python
def height_to_meters(feet, inches):
    total_inches = feet * 12 + inches
    return total_inches * 0.0254
```

```python
print(height_to_meters(6, 0))
print(height_to_meters(6, 2))   
```

```python
def meters_to_feet_and_inches(meters):
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches - (feet * 12)
    return feet, inches
```

```python
test_values = [2, 3, 4]

for m in test_values:
    ft, inch = meters_to_feet_and_inches(m)
    m_old = height_to_meters(ft, inch)
    diff = abs(m - m_old)

    print("original:", m, "old:", m_old)
```
